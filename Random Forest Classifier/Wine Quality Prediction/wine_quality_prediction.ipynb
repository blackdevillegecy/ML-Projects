#### Splitting data into Features and Target

```python
X = wine_quality_data.drop(columns="quality", axis=1)
Y = wine_quality_data["quality"]
```

```python
X
```

```python
Y
```

#### Linear Binarization

```python
Y = wine_quality_data["quality"].apply(lambda y_val: 1 if y_val>=7 else 0)
```

```python
X
```

```python
Y
```

#### Splitting data into Train and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=2)
```

#### Model Training

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, Y_train)
```

```python
Y_test_cap = clf.predict(X_test)
acc_test = accuracy_score(Y_test, Y_test_cap)
```

```python
print(acc_test)
```

```python
input_data = (7.5, 0.5, 0.36, 6.1, 0.071, 17.0, 102.0, 0.9978, 3.35, 0.8, 10.5)

arr = np.asarray(input_data)

arr_reshape = arr.reshape(-1, 1)

prediction = clf.predict(arr_reshape.T)

if prediction[0] == 1:
    print("Good Quality 7")
else:
    print("Less Quality 7")
```
