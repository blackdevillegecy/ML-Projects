#### Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
import warnings
warnings.filterwarnings("ignore")
```

#### Data Collection and Analysis

```python
student_data = pd.read_csv("student_marks.csv")
```

```python
student_data
```

#### Model Training

```python
r_state = np.random.RandomState(2)
```

```python
model = IsolationForest(n_estimators=100, random_state=r_state)
```

```python
model.fit(student_data[['marks']])
```

```python
student_data['f_s'] = model.decision_function(student_data[['marks']])
```

```python
student_data['score'] = model.predict(student_data[['marks']])
```
